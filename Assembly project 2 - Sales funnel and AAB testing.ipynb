### Посчет и визуализация воронки

Построим график с воронкой по всем пользователям с показателями потерь и долей пользователей 

```python
fig = go.Figure(go.Funnel(x = user_events['users'], y = user_events['event_name'], textinfo = "value+percent initial+percent previous"))
fig.show();
```

По построенному графику воронки можно увидеть, что значительная часть клиентов теряется между первым и вторым шагом (переход от главной страницы к странице с товарами) Рассчитаем процент отдельно позже. Пока рассмотрим графики по группам. Возможно, там мы сможем увидеть что-нибудь полезное для исследования.

Итак, настало время для проведения тестирования для проверки уровня статистической значимости. После всех проделанных итераций мы имеем: Обрезанные данные, в которых уникальные пользователи находятся в своих группах, пересечений в группах между пользователями - отсутствует, что дает нам понимание независимости выборок. Выборки между собой имеют незначительную разницу в объеме. <br>
Произведем z-тест для сравнения долей групп в каждом событии. Создадим функйию для проведения теста по каждому действую между контрольными группами.

### A1/A2 тест

Сформулируем гипотезы: <br>
H0: между долями нет значимой разницы <br> 
H1: между долями есть значимая разница <br>
Уровень статистической значимости - 0.01 Так как это является одним из условий проведения теста контрольных групп (А/А)

```python
def z_test(group1, group2, alpha):
    for i in event_group_test.index:
        p1 = event_group_test[group1][i] / users_bygroup[group1]
        # пропорция успехов во второй группе:
        p2 = event_group_test[group2][i] / users_bygroup[group2]
        # пропорция успехов в комбинированном датасете:
        p_combined = ((event_group_test[group1][i] + event_group_test[group2][i]) / 
                      (users_bygroup[group1] + users_bygroup[group2]))
        # разница пропорций в датасетах
        difference = p1 - p2
        # считаем статистику в ст.отклонениях стандартного нормального распределения
        z_value = difference / math.sqrt(p_combined * (1 - p_combined) * 
                                        (1/users_bygroup[group1] + 1/users_bygroup[group2]))
        # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
        distr = st.norm(0, 1) 
        p_value = (1 - distr.cdf(abs(z_value))) * 2
        print('{} p-значение: {}'.format(event_group_test['event_name'][i], p_value))
        if (p_value < alpha):
            print("Отвергаем нулевую гипотезу: между долями есть значимая разница")
        else:
            print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными")
        print('')
        
z_test("A1", "A2", 0.01)
```

По результатам проведения тестирования, по каждому из произведенных действий клиентов по группам доли между группами не являются статистически разными

Проведем тестирование между дургими группами. (А1/В), (А2/В) и (А1,2/В) <br>
Гипотезы остаются теми же. <br>
H0: между долями нет значимой разницы <br> 
H1: между долями есть значимая разница <br>
Здесь уровень статистической значимости можно принять как 0.05.
